```python
# converte wage em string para numero 
def extract_wages(x):
    out = x.replace("€", "")
    if "K" in out:
        out = float(out.replace("K", ""))*1000
    elif "M" in out:
        out = float(out.replace("M", ""))*1000000
    return float(out)
```

```python
df["Wage_Num"] = df["Wage"].apply(lambda x: extract_wages(x))
df.head()
```
